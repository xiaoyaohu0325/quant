```python
import pandas as pd
sz50 = pd.read_csv("sz50.csv")
sz50 = sz50.loc[:, ["date", "close"]]
sz50.set_index("date")
sz50.head()
```

```python
import numpy as np

def cal_avg(close_price, days):
    avg = np.zeros((len(close_price,)))
    avg[:days] = np.nan

    for i in range(days, len(avg)):
        avg[i] = np.sum(close_price[i-days:i]) / days

    return avg
```

```python
avg5 = cal_avg(sz50["close"], 5)
sz50["avg5"] = avg5
avg10 = cal_avg(sz50["close"], 10)
sz50["avg10"] = avg10
avg20 = cal_avg(sz50["close"], 20)
sz50["avg20"] = avg20
avg30 = cal_avg(sz50["close"], 30)
sz50["avg30"] = avg30
sz50.head(10)
```

```python
sz50.to_hdf("avg_index.h5", key="sz50", mode='a', complevel=4, complib="zlib")
```
